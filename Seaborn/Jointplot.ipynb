```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
sns.set_style("whitegrid")
```

```python
j1 = sns.jointplot(data=df, x='fare', y='age')
j2 = sns.jointplot(data=df, x='fare', y='age', kind='reg')
j3 = sns.jointplot(data=df, x='fare', y='age', kind='hex')
j4 = sns.jointplot(data=df, x='fare', y='age', kind='kde')

j1.fig.suptitle('titanic fare - scatter', size = 10)
j2.fig.suptitle('titanic fare - reg', size = 10)
j3.fig.suptitle('titanic fare - hex', size = 10)
j4.fig.suptitle('titanic fare - kde', size = 10)
```
