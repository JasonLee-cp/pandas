```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')

sns.set_style('darkgrid') # Total 5: darkgrid, whitegrid, dark, white, ticks

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

# Stripplot
sns.stripplot(data=df, x='class', y='age', ax = ax1)

# Swarmplot
sns.swarmplot(data=df, x='class', y='age', ax=ax2)

ax1.set_title('strip plot', c='red', size=20)
ax2.set_title('swarm plot', c='red', size=20)
```

```python
fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

# Stripplot
sns.stripplot(data=df, x='class', y='age', ax = ax1, hue='sex')

# Swarmplot
sns.swarmplot(data=df, x='class', y='age', ax=ax2, hue='sex')

ax1.set_title('strip plot', c='red', size=20)
ax2.set_title('swarm plot', c='red', size=20)
```
