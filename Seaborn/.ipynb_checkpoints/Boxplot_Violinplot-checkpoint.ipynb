```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Categorical Data

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
sns.set_style('darkgrid')
```

```python
fig = plt.figure(figsize=(15,10))

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

# Box plot
sns.boxplot(data=df, x='alive', y='age', ax = ax1)
sns.boxplot(data=df, x='alive', y='age', hue='sex', ax = ax2)
sns.violinplot(data=df, x='alive', y='age', ax = ax3)
sns.violinplot(data=df, x='alive', y='age', hue='sex', ax=ax4)

plt.show()
```
