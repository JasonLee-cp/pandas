```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../../../05000266/part4/auto-mpg.csv', header = None)
df.columns = ['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','name']
df
```

```python
df['count'] = 1
df
```

```python
df_origin = df.groupby('origin')['count'].sum()
```

```python
df_origin
```

```python
df_origin.index=['USA','EU','JPN']

df_origin.plot(kind='pie', figsize=(8,8), autopct='%1.1f%%',startangle=10, colors=['chocolate','bisque','cadetblue'])
plt.title('Model Origin', size = 20)
plt.axis('equal')
plt.legend(labels=df_origin.index, loc='best')
plt.show()
```
